```python
import pandas as pd
import numpy as np
```

Generate income for a non-defaulting individual

```python
fortnightly_income = np.random.randint(350, 520, 26)
reported_fortnightly_income = fortnightly_income
```

```python
df_income = pd.DataFrame()
df_income['actual_income'] = fortnightly_income
df_income['reported_income'] = reported_fortnightly_income
df_income['fortnights'] = [x + 1 for x in range(26)]
df_income.set_index('fortnights', inplace=True)
df_income
```

```python
df_income.to_csv('non_defaulter_incomes.csv')
```
